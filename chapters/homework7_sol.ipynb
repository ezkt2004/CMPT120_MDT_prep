## Practice Question 1

Write a function called `first_n(filename, n)` that prints the first `n` line of
the file `filename`, in the order they occur, and with line numbers as shown
below.

For example, the file [example.txt](example.txt) contains this:

```
The road to wisdom?—Well, it’s
plain and simple to express:
Err
and err
and err again,
but less
and less
and less.
-- Piet Hein
```

`first_n('example.txt', 3)` prints:

```
1. The road to wisdom?—Well, it’s
2. plain and simple to express:
3. Err
```

You can assume `n` is always a non-negative integer and that it is never more
than the number of lines in the file. This is not a practical assumption, but it
simplifies the problem. When you are done, think about how you might handle the
case where `n` is greater than the number of lines in the file.

*Don't* use lists or other such data structures in your answer.

````python
#
# sample solution
#

def first_n(filename, n):
    """Prints the first n lines of the file.
    Numbers each line. Assumes the file has at least n lines.
    """
    file_object = open(filename)
    for i in range(n):
        line = file_object.readline().strip()
        print(f'{i+1}. {line}')


first_n('example.txt', 3)
````

## Practice Question 2

Modify the code for the Spelling Bee game in the notes so that it also prints
the highest-scoring word. 

For example, for letters SIGLENP with required letter G, the output looks like
this:

```
... list of words ...

Total score: 1324
Total scoring words: 202
Best word: SLEEPINGS, 16 points
```

If the letters are LGXNCEI and the required letter is I, then:

```
... list of words ...

Total score: 401
Total scoring words: 72
Best word: EXCELLING, 16 points
```

If there's more than one highest-scoring word, print the one that comes first
alphabetically.

*Don't* use lists or other such data structures in your answer. Also, keep in
mind that [words.txt](words.txt) is already in alphabetical order.

````python
#
# Sample solution
#

def pangram_bonus(word, letters):
    """Returns 7 if every character in letters is in words.
    Otherwise it returns 0.
    """
    for c in letters:
        if c not in word:
            return 0
    return 7


def word_score(word, required, letters):
    """Returns the score for word according to these rules:
    - Words with 3 letters aren't considered (they don't score any points).
    - A 4-letter word is worth 1 point.
    - A 5-letter or more word is worth 1 point per letter.
    - A word that uses all seven letters is called a *pangram* and is worth 7
      extra points.
    """
    #
    # first check if the word is too short or doesn't contain the required
    # letter
    #
    if required not in word or len(word) < 4:
        return 0
    
    #
    # check that all the letters in word are in letters
    #
    for c in word:
        if c not in letters:
            return 0
    
    #
    # if we get here, the word is long enough and contains the required letter
    #
    if len(word) == 4:
        return 1
    else:
        return len(word) + pangram_bonus(word, letters)

# sample
# letters = 'SIGLENP'
# required_letter = 'G'

# June 20, 2024
letters = 'LGXNCEI'
required_letter = 'I'

best_word = ''
best_score = 0
total_score = 0    # sum of all the scores
scoring_words = 0  # number of words that score points
file_object = open('words.txt')
for w in file_object:
    # remove the newline character at the end of the word
    # also cover it to upper case
    w = w.strip().upper()
    
    score = word_score(w, required_letter, letters)
    total_score += score
    if score > 0:
        scoring_words += 1
        if score > best_score:
            best_word = w
            best_score = score
        print(f'{w}, {score} points')

print()
print(f'Total score: {total_score}')
print(f'Total scoring words: {scoring_words}')
best_score = word_score(best_word, required_letter, letters)
print(f'Best word: {best_word}, {best_score} points')
````

## Practice Question 3

Write a function that prints all the words in [words.txt](words.txt) that match
this rule:

- contains two or more *x*s (not necessarily consecutive)

The output should look like this:

```
executrix has 2 x's
executrixes has 2 x's
exotoxic has 2 x's
exotoxin has 2 x's
exotoxins has 2 x's
maxixe has 2 x's
maxixes has 2 x's
paxwax has 2 x's
paxwaxes has 2 x's
9 words have two or more x's
```

Of course, your function should word for any list of words, not just the ones in
[words.txt](words.txt).

````python
#
# Sample solution
#

def print_two_or_xs():
    count = 0
    file_object = open('words.txt')
    for w in file_object:
        w = w.strip()
        x_count = w.count('x')
        if x_count >= 2:
            print(f'{w} has {x_count} x\'s')
            count += 1
    print(f'{count} words have two or more x\'s')

print_two_or_xs()
````

## Practice Question 4

Write a function that prints all the words in [words.txt](words.txt) that match
this rule:

- contains 6 or more copies of the same letter, other than *s*

The output should look like this:

```
executrix has 2 x's
executrixes has 2 x's
exotoxic has 2 x's
exotoxin has 2 x's
exotoxins has 2 x's
maxixe has 2 x's
maxixes has 2 x's
paxwax has 2 x's
paxwaxes has 2 x's
9 words have two or more x's
```

Of course, your function should word for any list of words, not just the ones in
[words.txt](words.txt).

````python
#
# Sample solution
#
def print_six_or_more_letters():
    count = 0
    file_object = open('words.txt')
    for w in file_object:
        w = w.strip()
        for c in 'abcdefghijklmnopqrstuvwxyz':
            letter_count = w.count(c)
            if c != 's' and letter_count >= 6:
                print(f'{w}, {c} appears {letter_count} times')
                count += 1
    print(f'{count} words have a letter that occurs at least 6 times')

print_six_or_more_letters()
````

## Practice Question 5

Write a function that prints then number of times *a*, *e*, *i*, *o*, and *u*
occur in in [words.txt](words.txt). Print a summary at the end.

For instance, in the words *doodle*, *e* occurs 2 times, and `e` occurs 1 time.
The result of the vowels appear 0 times.

The output should look like this:

```
a: 68574
e: 106752
i: 77392
o: 54538
u: 31151
```

Of course, your function should word for any list of words, not just the ones in
[words.txt](words.txt).

````python
#
# Sample solution
#
def print_vowel_counts():
    file_object = open('words.txt')
    a_count = 0
    e_count = 0
    i_count = 0
    o_count = 0
    u_count = 0
    for w in file_object:
        w = w.strip()
        a_count += w.count('a')
        e_count += w.count('e')
        i_count += w.count('i')
        o_count += w.count('o')
        u_count += w.count('u')
    print(f'a: {a_count}\ne: {e_count}\ni: {i_count}\no: {o_count}\nu: {u_count}')

print_vowel_counts()
````

## Practice Question 7

Write a function called `print_file_sum(fname)` that reads the sum of the
numbers in the file `fname`. Assume that `fname` is a valid file name.

For example, calling these functions:

```python
print_file_sum('empty.txt')
print_file_sum('num1.txt')
print_file_sum('num2.txt')
print_file_sum('num5.txt')
```

should print:

```
Sum of numbers in empty.txt: 0.0
Sum of numbers in num1.txt: 3.5
Sum of numbers in num2.txt: 5.35
Sum of numbers in num5.txt: 6.31
```

````python
#
# Sample solution
#
def print_file_sum(fname):
    file_obj = open(fname)
    total = 0
    for line in file_obj:
        total += float(line)
    print(f'Sum of numbers in {fname}: {total}')

print_file_sum('empty.txt')
print_file_sum('num1.txt')
print_file_sum('num2.txt')
print_file_sum('num5.txt')
````

## Question 8

Write a function called `print_file_stats(fname)` that prints the average and
the [standard deviation](https://en.wikipedia.org/wiki/Standard_deviation) of
the numbers in the file `fname`.

If the $n$ numbers in the file are called $x_1, x_2, \ldots, x_n$, then the
average is:

$$
\text{avg} = \frac{x_1 + x_2 + \cdots + x_n}{n}
$$

And the standard deviation is:

$$
\text{standard deviation} = \sqrt{\frac{(x_1-\text{avg})^2}{n} + \frac{(x_2-\text{avg})^2}{n} + \ldots + \frac{(x_n-\text{avg})^2}{n}}
$$

If you call:

```python
print_file_stats('empty.txt')
print_file_stats('num1.txt')
print_file_stats('num2.txt')
print_file_stats('num5.txt')
```

Your program should print this:

```
empty.txt error: must contain at least 2 numbers (has 0)

num1.txt error: must contain at least 2 numbers (has 1)

              File: num2.txt
           Average: 2.67
Standard deviation: 0.47

              File: num5.txt
           Average: 1.26
Standard deviation: 2.29
```

**Hint** If you have looped through the files once already, you can call
`file_obj.seek(0)` to reset the file object so that if you loop through it again
it will start from the beginning.

````python
#
# Sample solution
#
import math # for sqrt

def print_file_stats(fname):
    #
    # get the average
    #
    file_obj = open(fname)
    total = 0
    count = 0
    for line in file_obj:
        total += float(line)
        count += 1
    
    #
    # if there are no numbers, print a message and return
    #
    if count < 2:
        print(f'{fname} error: must contain at least 2 numbers (has {count})')
        return
    
    # if we get here, there is at least one number in the file
    avg = total / count
    
    #
    # get the standard deviation
    #
    # file_obj.close()
    # file_obj = open(fname)
    
    # reset the file object to the beginning
    file_obj.seek(0)
    
    sd_total = 0
    for line in file_obj:
        sd_total += (float(line) - avg) ** 2
    std_dev = math.sqrt(sd_total / count)
    
    print(f'              File: {fname}')
    print(f'           Average: {avg:.2f}')
    print(f'Standard deviation: {std_dev:.2f}')

print_file_stats('empty.txt')
print()
print_file_stats('num1.txt')
print()
print_file_stats('num2.txt')
print()
print_file_stats('num5.txt')
````
