## Function Return Values

Functions like `abs` and `sqrt` **return** a value that you can assign to a
variable or use in an expression. But some functions, like `print` and
`turtle.forward`, *don't* return a value.

To write your own function that returns a value, the function must end with a
call to `return`:

````python
import math

def circle_area(radius):
    area = math.pi * radius**2
    return area  # return statement
````

The last line of `circle_area` is a **return statement**. Python return
statements always start with the keyword `return`.

We can save the value it returns in a variable:

````python
r = float(input('What is the radius? '))

if r <= 0:
    print('radius must be greater than 0')
else:
    area = circle_area(r)
    print(f'A circle of radius {r} has area {area}.')
````

## Functions without a Return

In Python, functions that don't return a value return the special value `None`. For instance:

````python
def greet(name):
    print(f'Hello {name}!')

greet('Elon')
````

Look what happens if you assign `greet` to a variable:

````python
x = greet('Elon')   # prints Hello Elon!
print(x)            # None
````

`x` gets the special value `None`. Python uses `None` to mean that a function doesn't return a value. If you get `None` values in your program, check for missing `return` statements, or for places you may be accidentally using a non-returning function to a variable.

We could have written `greet` to return a value:

````python
def greet_result(name):
    return f'Hello {name}!'

greet_result('Elon')      # does nothing

x = greet_result('Elon')  # x is 'Hello Elon!'
print(x)                  # prints Hello Elon!
````

## Return Values and Conditionals

Python's `abs(x)` function could be written like this:

````python
def abs_val(x):
    if x < 0:
        return x
    else:
        return -x

print(abs_val(-2))  # 2
print(abs_val(0))   # 0
print(abs_val(2))   # 2
````

There are two `return` statements, but only one is ever executed on any call to
`abs_val`.

We say there are two **paths** through the program, one ending with `return x`
and one ending with `return -x`. When a function with multiple paths returns a
value, there must be a `return` at the end of every possible path through the
function.

Here's an example of an incorrect implementation of absolute value:

````python
def bad_abs_val(x):
    if x < 0:
        return -x
    if x > 0:
        return x

print(bad_abs_val(-2))  # ok, 2
print(bad_abs_val(2))   # ok, 2
print(bad_abs_val(0))   # error, None
````

In `bad_abs_val(x)` there is no `return` at the end of the path taken when `x` is 0. So `bad_abs_val(0)` gives the special value `None`, indicating no value was returned.

It's possible to have extra `return` statements. For example:

````python
def abs_val_extra_return(x):
    if x < 0:
        return -x
    else:
        return x
    
    return 'This is dead code'  # dead code: never run
````

The `return` statement at the bottom is never run because the function always ends by calling one of the `return` statements above it. Code that is never run is called **dead code**, and it is often a sign of a logic bug, or misunderstanding.

Here are a couple of other examples of dead code:

````python
def dead_code_ex1():
    return 1
    return 2  # dead code
    return 3  # dead code

def dead_code_ex2():
    n = 0
    for i in range(n):
        print('Hello!')  # dead code

def dead_code_ex3():
    if x <= 0:
        print('up')
    elif x >= 0:
        print('down')
    else:
        print('around')  # dead code

def dead_code_ex4():
    dead_code_ex4()
    print('done!')  # dead code
````

## Boolean Functions

A function that returns either `True` or `False` is called a **boolean function**. For example:

````python
def is_positive(x):
    if x > 0:
        return True
    else:
        return False

print(is_positive(-2))  # False
print(is_positive(0))   # False
print(is_positive(2))   # True
````

This works fine, but there is a shorter way to write this function:

````python
def is_positive(x):
    return x > 0

print(is_positive(-2))  # False
print(is_positive(0))   # False
print(is_positive(2))   # True
````

The expression `x > 0` returns either `True` or `False`, and so we don't need to
use an if-statement. Most programmers prefer this second way because it is short
and readable (at least once you get used to it!).

### Example: Testing if a Number is Prime

A **prime number** is an integer greater than 0 number that is only divisible by
1 and itself. For example, 2 is the first prime number (and the only even prime
number), and the next few primes are 3, 5, 7, 11, 13, ....

Let's write a function `is_prime(n)` that returns `True` if `n` is a prime
number and `False` otherwise. We'll assume that `n` is an `int`.

First, we can handle a couple of special cases:
- if `n` is less than 2, then it is *not* prime.
- if `n` is 2, then it *is* prime.
- if `n` is even and greater than 2, then it is *not* prime.

So are function looks like this:

```python
def is_prime(n):
    """Returns True if n is a prime number, False otherwise.
    """
    if n < 2:
        return False
    elif n == 2:
        return True
    elif n % 2 == 0:
        return False
    else:
        # n > 2 and n is odd
```

For the `else` part, we know that `n` is bigger than 2 and is odd. If none of
the numbers 2, 3, 4, ..., `n-1` evenly divide `n`, then `n` is prime. Otherwise,
if one, or more of 2, 3, 4, ..., `n-` evenly divide `n`, then `n` is *not*
prime.

So we can use a loop to generate all the numbers from 2 to `n-1` and check if
any of them evenly divide `n`. We also write a few test cases to check that it
works correctly:

````python
def is_prime(n):
    """Returns True if n is a prime number, False otherwise.
    """
    if n < 2:
        return False
    elif n == 2:
        return True
    elif n % 2 == 0:
        return False
    else:
        # n > 2 and n is odd
        for i in range(3, n):
            if n % i == 0:
                return False
        return True

# test code
for i in range(-2,20):
    if is_prime(i):
        print(f'{i} is prime')
    else:
        print(f'{i} is NOT prime')
````

It appears that the function works correctly: it has passed all the test cases.
Of course, it is possible there is still a bug in the function, but there is no
way to be certain from test cases alone. In addition to the test cases, we can
read the function and reason about what it is doing, and see that it appears to
be correct.

The function has a couple of inefficiencies:

- If `n` is *not* prime, then it must have a factor less than, or equal to, its
  square root. So checking all the way to `n-1` is not necessary.
- Since `n` is odd in the `else` part, we know that `n` is not divisible by 2.
  So we don't need to check if any even number divides `n`.

So lets add those improvements to the function:

````python
def is_prime2(n):
    """Returns True if n is a prime number, False otherwise.
    """
    if n < 2:
        return False
    elif n == 2:
        return True
    elif n % 2 == 0:
        return False
    else:
        # n > 2 and n is odd
        square_root = int(n**0.5)
        for i in range(3, square_root + 1, 2):
            if n % i == 0:
                return False
        return True
    
# test code
for i in range(-2,20):
    if is_prime2(i):
        print(f'{i} is prime')
    else:
        print(f'{i} is NOT prime')
````

The for-loop `range` now starts at 3 and goes up to the square root of `n`, and
it adds 2 to `i` each time. Since we start at 3, this means we only check the
odd numbers 3, 5, 7, etc.

### Example: The Prime Counting Function

[The prime counting
function](https://en.wikipedia.org/wiki/Prime-counting_function), denoted
$\pi(n)$, is the number of prime numbers less than or equal to $n$. For example,
$\pi(10) = 4$ because there are four prime numbers less than or equal to 10: 2,
3, 5, and 7.

We can write a function `num_primes(n)` that returns the number of prime numbers
less than, or equal to, `n` using the `is_prime` function:

````python
def num_primes(n):
    """Returns the number of primes less than, or equal to, n.
    """
    count = 0
    for i in range(2, n + 1):
        if is_prime(i):
            count += 1
    return count

print(num_primes(10))  # 4
print(num_primes(20))    # 8
print(num_primes(100))   # 25
print(num_primes(1000))  # 168  
````

## Recursive Functions with Return Values

Since recursive functions are functions, they can have return values. 

For example, suppose we want to calculate the sum $1 + 2 + 3 + \ldots + n$. You
may recall from math class that $1 + 2 + 3 + \ldots + n = \frac{n(n+1)}{2}$. So
we could write a function that calculates the sum like this:

````python
def sum_to(n):
    return n * (n + 1) // 2

print(sum_to(1))  #  1 = 1
print(sum_to(2))  #  3 = 1 + 2
print(sum_to(3))  #  6 = 1 + 2 + 3
print(sum_to(4))  # 10 = 1 + 2 + 3 + 4
````

This is a good and efficient way to calculate $1 + 2 + 3 + \ldots + n$.

Now lets re-implement the function using recursion. To simplify this, first lets
define $S(n)$ to be this:
$$
S(n) = 1 + 2 + 3 + \ldots + n
$$

For example, $S(4) = 1 + 2 + 3 + 4 = 10$.

A problem with this definition is that `...` does not have a precise enough
meaning for a computer. While humans know that `...` means something like
"continue in the same pattern", a programming language Python cannot recognize
such a pattern.

So here is a more precise definition:
$$
\begin{aligned}
S(0) &= 0          & \textrm{base case} \\
S(n) &= n + S(n-1) & \textrm{recursive case}  \\
\end{aligned}
$$

This is a **recursive definition** of $S(n)$, also known as a **recurrence
relation**. It defines $S(n)$ precisely enough that we can write a function to
calculate it.

````python
def sum_rec(n):
    if n == 0: # base case
        return 0
    else:      # recursive case
        return n + sum_rec(n - 1)

print(sum_rec(1))  #  1 = 1
print(sum_rec(2))  #  3 = 1 + 2
print(sum_rec(3))  #  6 = 1 + 2 + 3
print(sum_rec(4))  # 10 = 1 + 2 + 3 + 4
````

Recursion can look like magic at first. It helps to trace through an example by
hand to see what is going on. Let's calculate $S(4)$ using the recursive
formula. To calculate $S(4)$, the formula says to calculate $4 + S(3)$. We don't
know what $S(3)$ is, so we use the formula to calculate $S(3)$, which is $S(2)+
3$. We don't know what $S(2)$ is, so again we use the formula to calculate
$S(2)$, which is $S(1) + 2$. We repeatedly apply the recursive formula like this
until we reach the base case of $S(0)$, which is just 0. 

Step-by-step it looks like this:

$$
\begin{aligned}
S(4) &= 4 + S(3) \\
     &= 4 + 3 + S(2) \\
     &= 4 + 3 + 2 + S(1) \\
     &= 4 + 3 + 2 + 1 + S(0) \\
     &= 4 + 3 + 2 + 1 + 0 \\
     &= 10
\end{aligned}
$$

As mentioned, this is not an efficient way to calculate $1 + 2 + \ldots + n$,
but is a good example of recursion.

## Fibonacci Numbers

Let's look at another recursive definition, the [Fibonacci
numbers](https://en.wikipedia.org/wiki/Fibonacci_number). The Fibonacci numbers
are the sequence $0, 1, 1, 2, 3, 5, 8, 13, 21, \ldots$. The first two numbers
are 0 and 1, and each following number is the sum of the two before it.

We can define the numbers like this:

$$\begin{aligned}
\mathrm{fib}(0) &= 0 & \textrm{base case 1} \\
\mathrm{fib}(1) &= 1 & \textrm{base case 2} \\
\mathrm{fib}(n) &= \mathrm{fib}(n-1) + \mathrm{fib}(n-2) & \textrm{recursive case}
\end{aligned}$$ 

This has two base cases: $\mathrm{fib}(0) = 0$ and $\mathrm{fib}(1) = 1$. The
recursive case is interesting since it calls itself *twice*, once with $n-1$ and
once with $n-2$

Translating this into Python is straightforward:

````python
def fib(n):
    """Returns the nth Fibonacci number.
    fib(0) = 0, fib(1) = 1, fib(2) = 1, ...
    """
    if n == 0:   # base case 1
        return 0
    elif n == 1: # base case 2
        return 1
    else:        # recursive case
        return fib(n - 1) + fib(n - 2)

print(fib(0))  # 0 
print(fib(1))  # 1
print(fib(2))  # 1
print(fib(3))  # 2
print(fib(4))  # 3
print(fib(5))  # 5
print(fib(6))  # 8
````

Notice that `fib(-1)` causes the function to crash. This is because the
recurrence relation is only defined for $n \geq 0$. Calling `fib` with any value
less than 0 will make it crash.

This function has another major problem: it is *extremely* inefficient for
anything but small values of `n`. This is because it recalculates the same
values over and over again.

For instance, `fib(40)` takes over 20 seconds to calculate on my computer:

````python
print(fib(40))  # 102334155, over 20 seconds!
````

So do **not** use this as a way to calculate Fibonacci numbers in practice!
